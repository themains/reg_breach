### Get All FL Voter Reg. Emails

```python
import os
import glob
import pandas as pd
```

```python
# Specify the directory path containing the CSV files
directory_path = 'data/VoterDetail/'

# Get a list of all CSV file paths in the directory
csv_files = glob.glob(os.path.join(directory_path, '*.txt'))

dataframes = []

for fn in sorted(csv_files):
    print(fn)
    a_df = pd.read_csv(fn, delimiter='\t', header=None, usecols=[2, 4, 37])
    a_df.dropna(inplace = True)
    dataframes.append(a_df)

df = pd.concat(dataframes, ignore_index=True)
df
```

```python
import numpy as np
df.columns = ['lname', 'fname', 'email']
df['email'] = df['email'].str.strip()
df['email'].replace('', np.nan, inplace=True)
df.dropna(subset = ['email'], inplace = True)
df
```

```python
df.to_csv("fl_emails.csv", index = False)
```
