### Get All FL Voter Reg. Emails

```python
import os
import glob
import pandas as pd
import numpy as np
```

```python
# Specify the directory path containing the CSV files
directory_path = '../data/VoterDetail/'

# Get a list of all CSV file paths in the directory
csv_files = glob.glob(os.path.join(directory_path, '*.txt'))

dataframes = []

for fn in sorted(csv_files):
    print(fn)
    a_df = pd.read_csv(fn, delimiter='\t', header=None, usecols=[17, 19, 20, 21, 37])
    dataframes.append(a_df)

df = pd.concat(dataframes, ignore_index=True)
df
```

```python
df.rename(columns = {17: 'zipcode', 
           19: 'gender', 
           20: 'race', 
           21: 'birth_yr', 
           37: 'email'}, inplace = True)
df['email'] = df['email'].str.strip()
df['email'].replace('', np.nan, inplace=True)
df.dropna(subset = ['email'], inplace = True)
df
```

```python
# Recode values in column 'A'
recode_map = {1: 'Native Americans', 
              2: 'Asian', 
              3: 'NH Black', 
              4: 'Hispanic', 
              5: 'NH White', 
              6: 'Other', 
              7: 'Multi-Racial', 
              9: 'Unknown'}
df['race_lit'] = df['race'].replace(recode_map)
df['race_lit'].value_counts()
```

```python
df.to_csv("fl_emails.csv", index = False)
```
